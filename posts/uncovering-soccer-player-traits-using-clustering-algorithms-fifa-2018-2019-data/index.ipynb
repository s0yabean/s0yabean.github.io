# Introduction (What Is Clustering?)

In the realms of machine learning, a very common delineation is between:

- supervised learning 
- unsupervised learning

The key difference is whether a target variable of interest is available. In other words, is there something for the training data to predict and compare with? 

If not, it typically is an unsupervised learning problem, where we seek to unravel the "hidden" target variable using techniques that attempt to group similar data together.

# What Is The Goal?

In the big leagues, teams are constantly in flux:

- Players are bought and sold (2019 EPL transfer window value was £1.41bn)
- Players get injured or ill (Average injuries per team in EPL 16/17 season was 57.5. The maximum team size is 25)
- Player's mentality/ form/ physical performance changes over time 

However, coaches aspire to have consistent performances as much as possible, pertaining to a specific style or end-result. 

Top teams have a distinct style - Jurgen Klopp is known for his ‘Gegenpressing’. The EPL is known for being physical and direct, emanating the swashbuckling style of the English. How about the "Tiki-Taka" philosophy made famous by Pep Guardiola?

Coaches work with scouts to identify new players with similar profiles to their current squad.

This prepares them for ready substitutions - in case for sudden injuries, downtime or other reasons that leave a gaping hole in their formations, they still have someone to play a specific role in the team.

Hence, I will be building a tool for coaches and scouts that would recommend similar players outside their current squad to them, so they can acquire the right "replacements" for their desired roles, whoever the player is.
